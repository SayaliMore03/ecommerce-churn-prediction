```python
import pandas as pd
import numpy as np

df = pd.read_excel('../data/raw/E Commerce Dataset.xlsx', sheet_name='E Comm')

print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
df.head()
```

```python
missing = df.isnull().sum()
missing_pct = (df.isnull().sum() / len(df) * 100).round(2)

missing_df = pd.DataFrame({
    'Missing Count': missing,
    'Missing %': missing_pct
})

print(missing_df[missing_df['Missing Count'] > 0])
```

```python
print("Duplicate rows:", df.duplicated().sum())
print("Duplicate CustomerIDs:", df['CustomerID'].duplicated().sum())
```

```python
for col in df.select_dtypes(include='object').columns:
    print(f"\n{col}: {sorted(df[col].unique())}")
```

```python
print("CashbackAmount nulls:", df['CashbackAmount'].isnull().sum())
```

```python
df = df.rename(columns={'PreferedOrderCat': 'PreferredOrderCat'})
print("Renamed ✓ New columns:", df.columns.tolist())
```

```python
df['PreferredLoginDevice'] = df['PreferredLoginDevice'].replace({
    'Mobile Phone': 'Mobile',
    'Phone': 'Mobile'
})

df['PreferredPaymentMode'] = df['PreferredPaymentMode'].replace({
    'CC': 'Credit Card',
    'COD': 'Cash on Delivery'
})

df['PreferredOrderCat'] = df['PreferredOrderCat'].replace({
    'Mobile Phone': 'Mobile'
})

# Verify all fixed
for col in ['PreferredLoginDevice', 'PreferredPaymentMode', 'PreferredOrderCat']:
    print(f"{col}: {sorted(df[col].unique())}")
```

```python
cols_to_fill = [
    'Tenure', 'WarehouseToHome', 'HourSpendOnApp',
    'OrderAmountHikeFromlastYear', 'CouponUsed',
    'OrderCount', 'DaySinceLastOrder', 'CashbackAmount'
]

for col in cols_to_fill:
    median_val = df[col].median()
    df[col] = df[col].fillna(median_val)
    print(f"{col}: filled with median = {median_val}")
```

```python
int_cols = ['CouponUsed', 'OrderCount', 'DaySinceLastOrder']
df[int_cols] = df[int_cols].astype(int)

print("Updated dtypes:")
print(df[int_cols].dtypes)
```

```python
print("Nulls remaining:", df.isnull().sum().sum())
print("Shape:", df.shape)
print("Churn rate:", df['Churn'].mean().round(3))

# Range checks
assert df['SatisfactionScore'].between(1, 5).all(), "Bad satisfaction scores!"
assert df['Churn'].isin([0, 1]).all(), "Bad churn values!"
assert df['CityTier'].isin([1, 2, 3]).all(), "Bad city tier!"
assert (df['Tenure'] >= 0).all(), "Negative tenure!"

print("All validation checks passed ✓")
```

```python
df.to_csv('../data/processed/churn_clean.csv', index=False)
print("Clean data saved ✓")
print("Final shape:", df.shape)
print("Nulls remaining:", df.isnull().sum().sum())
```
